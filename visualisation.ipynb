```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
time_data = pd.DataFrame({'type': ['singlethreading']*4 + ['multithreading']*4 + ['multiprocessing']*4,
                          'seq_quantity': ['100', '1000', '10000', '300K']*3,
                          'time': [0.019, 0.055, 0.423, 9.534, 0.043, 0.103, 0.599, 15.890, 0.565, 0.661, 0.723, 3.736]})
```

```python
time_data
```

```python
sns.barplot(data=time_data.query('seq_quantity != "300K"'), x='seq_quantity', y='time', hue='type');
```

```python
proc_data = pd.DataFrame({'type': ['multithreading']*10 + ['multiprocessing']*10,
                          'n_workers': [i for i in range(1, 11)]*2,
                          'time': [15.185, 16.935, 16.584, 17.045, 17.993, 22.729, 22.330, 23.193, 21.956, 22.761,
                                   10.146, 5.842, 5.188, 4.684, 4.435, 4.291, 4.565, 4.302, 4.397, 4.363]})
```

```python
proc_data
```

```python
sns.barplot(data=proc_data, x='n_workers', y='time', hue='type');
```
